#### 인구통계자료 재가공

출처 : 부산광역시 공공데이터포털

```python
import os
import pandas as pd
root = '/workspace/data'
human_file_name = '인구통계자료/myfile.csv'

df_human = pd.read_csv(os.path.join(root, human_file_name), encoding='cp949') # csv 파일을 데이터프레임으로 불러옴
display(df_human)
```

```python
# 각 구와 부산광역시의 가장 최신 자료(Oct-22)만 남기기
df_human_oct_22 = df_human[df_human['제공연도'] == 'Oct-22']
display(df_human_oct_22)
```

```python
# 필요한 열만 남기기
df_human_oct_22_ = df_human_oct_22.loc[:, ['구군', '총인구수', '남성인구수', '여성인구수']]
df_human_oct_22_ = df_human_oct_22_.reset_index()
display(df_human_oct_22_)
```

```python
# mongodb에 저장
human_collection = db['human']
for k, v in df_human_oct_22_.iterrows():
    if k == 'index': continue
    data = {
        'seq': k,
        'gugun': v[1],
        'total': v[2],
        'total_male': v[3],
        'total_femal': v[4]
    }
    human_collection.insert_one(data)
```

```python
# 내용 확인
human_data = list(db.human.find({}, {'_id':False}))
for d in human_data:
    print(d)
```

#### 도서관 데이터 가공

- 지도 API를 활용하여 도서관 이름으로부터 도서관의 정확한 위치와 관할 시군구를 획득
- 도서관에 관한 데이터를 MongoDB에 저장

```python
# 공공데이터포털로부터 도서관 데이터 요청
import json
import requests

library_collection = db['library']

base_path = '/workspace/src'
with open(os.path.join(base_path, 'secret.json')) as f:
    secret_key = json.loads(f.read())

api_url = f"http://apis.data.go.kr/6260000/BusanLibraryInfoService/getLibraryInfo?ServiceKey={secret_key['library_information_key']}&numOfRows=50&pageNo=1&resultType=json"
response = requests.get(api_url)
contents = response.text
contents_json = json.loads(contents)

body = contents_json['getLibraryInfo']['body']['items']['item']
print(body)
```

```python
# 데이터프레임화
df_library = pd.json_normalize(body)
display(df_library)
```

```python
# MongoDB에 데이터 삽입

for k, v in df_library.iterrows():
    data = {
        'seq': k,
        'area': v[0],
        'library_name': v[1],
        'address': v[2],
        'tellephone': v[3],
        'site_url': v[4]
    }
    
    library_collection.insert_one(data)
```

```python
# 내용 확인
lib_data = list(db.library.find({}, {'_id':False}))
for d in lib_data:
    print(d)
```
