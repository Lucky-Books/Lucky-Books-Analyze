```python
import os
import pandas as pd

root = '/workspace/data/도서대출자료'
month = '1월'
book_file_name = 'january.csv'

df_book = pd.read_csv(os.path.join(root, month, book_file_name))
#df_book = df_book.loc[:, ['도서명', '저자', '출판사', '권', '도서권수', '대출건수', 'ISBN', '등록일자']]
df_book = df_book.loc[:, ['도서명', '대출건수']]
display(df_book)
```

```python
# min, max
df_book_min = df_book['대출건수'].min()
df_book_max = df_book['대출건수'].max()
print(f"최소 대출건수: {df_book_min}")
print(f"최대 대출건수: {df_book_max}")
```

```python
min_loan_books = df_book[df_book['대출건수'] == df_book_min]
max_loan_books = df_book[df_book['대출건수'] == df_book_max]

display(min_loan_books)
display(max_loan_books)
```
