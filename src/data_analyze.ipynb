```python
from IPython.display import display, HTML
import pandas as pd
from pymongo import MongoClient
```

```python
client = MongoClient('localhost', 27017)
db = client.luckybooks
```

```python
from pymongoarrow.api import Schema
from pymongoarrow.monkey import patch_all
import pyarrow as pa

patch_all()

schema = Schema({
    'seq': pa.int64(),
    'area': pa.string(),
    'library_name': pa.string(),
    'address': pa.string(),
    'tellephone': pa.string(),
    'site_url': pa.large_string()
})

library = db.library.find_pandas_all({}, schema=schema)
display(library)
```

```python
schema = Schema({
    'seq': pa.int64(),
    'gugun': pa.string(),
    'total': pa.int64(),
    'total_male': pa.int64(),
    'total_female': pa.int64()
})

human = db.human.find_pandas_all({}, schema=schema)
display(human)
```

```python
import os
root = "/workspace/data"
folder = "도서대출자료/1월"
l = os.listdir(os.path.join(root, folder))
print(l)
```

```python
df_jan = pd.DataFrame()
for i in l:
    book_file = os.path.join(root, folder, i)
    lib_name = ""
    for bf in book_file.split(' '):
        if "도서관" in bf:
            lib_name = bf
            if "/" in lib_name:
                lib_name = lib_name.split('/')[-1]
            if "부산광역시립" in lib_name:
                lib_name = lib_name.split('부산광역시립')[-1]
    df_bf = pd.read_excel(book_file)
    df_bf['도서관'] = lib_name
    df_jan = pd.concat([df_jan, df_bf])

display(df_jan)
```

```python
df_jan.to_csv(os.path.join(root, folder, 'january.csv'))
```
