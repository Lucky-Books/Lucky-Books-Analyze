#### 인구통계자료 재가공

출처 : 부산광역시 공공데이터포털

```python
import os
root = '/workspace/data'
human_file_name = '인구통계자료/myfile.csv'

df_human = pd.read_csv(os.path.join(root, human_file_name), encoding='cp949') # csv 파일을 데이터프레임으로 불러옴
display(df_human)
```

```python
# 각 구와 부산광역시의 가장 최신 자료(Oct-22)만 남기기
df_human_oct_22 = df_human[df_human['제공연도'] == 'Oct-22']
display(df_human_oct_22)
```

```python
# 필요한 열만 남기기
df_human_oct_22_ = df_human_oct_22.loc[:, ['구군', '총인구수', '남성인구수', '여성인구수']]
df_human_oct_22_ = df_human_oct_22_.reset_index()
display(df_human_oct_22_)
```

```python
# mongodb에 저장
human_collection = db['human']
for k, v in df_human_oct_22_.iterrows():
    if k == 'index': continue
    data = {
        'seq': k,
        'gugun': v[1],
        'total': v[2],
        'total_male': v[3],
        'total_femal': v[4]
    }
    human_collection.insert_one(data)
```

```python
# 내용 확인
human_data = list(db.human.find({}, {'_id':False}))
for d in human_data:
    print(d)
```
